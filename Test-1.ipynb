```python
def problem1(example, rotational_value):
    #To Encode the sentence i translate the character values to ascii using the ord command
    #I then add the rv value (positive or negative) to the ascii number and translate it back to
    #a character code.  As I am looping through the individual characters I am rebuilding the string
    #using concatenation and I then return the Encoded sentence.
    def Encode(sentence, rv):
        translated = ""
        for x in sentence:
            translated += chr(ord(x)+rv)
        return translated

    #To Decode I do the exact same formula as the Encode except i subtract the rotational value
    #This decodes the sentence an I return it.
    def Decode(sentence, rv):
        translated = ""
        for x in sentence:
            translated += chr(ord(x)-rv)
        return translated
    encoded = Encode(example, rotational_value)
    decoded = Decode(encoded, rotational_value)
    print("Enter a Message to Encrypt:", example)
    print("The Encrypted Message is:",encoded)
    print("The Decrypted Message is:",decoded)
problem1("Hello World!",13)
```

```python
import math
def flowercount(ls, flower_dict, H):
    for y in H:
        for x in ls:
            temp=y+"count"
            try:
                flower_dict[x[-1]][temp]+=1
            except:
                try:
                    flower_dict[x[-1]].update({temp:1})
                except:
                    flower_dict[x[-1]]={temp:1}
    return flower_dict
    
def flowersums(ls, flower_dict, H):
    z=-1
    for y in H:
        z+=1
        for x in ls:
            temp=y+"sum"
            try:
                flower_dict[x[-1]][temp]+=float(x[z])
            except:
                flower_dict[x[-1]].update({temp:float(x[z])})
    return flower_dict

def flowermins(ls, flower_dict, H):
    z=-1
    for y in H:
        z+=1
        for x in ls:
            temp=y+"min"
            try:
                if flower_dict[x[-1]][temp]>float(x[z]):
                    flower_dict[x[-1]][temp]=float(x[z])
            except:
                flower_dict[x[-1]].update({temp:float(x[z])})
    return flower_dict

def flowermax(ls, flower_dict, H):
    z=-1
    for y in H:
        z+=1
        for x in ls:
            temp=y+"max"
            try:
                if flower_dict[x[-1]][temp]>float(x[z]):
                    flower_dict[x[-1]][temp]=float(x[z])
            except:
                flower_dict[x[-1]].update({temp:float(x[z])})
    return flower_dict

def flowermean(ls, flower_dict, H):
    z=-1
    for y in H:
        z+=1
        for x in ls:
            temp=y+"mean"
            try:
                flower_dict[x[-1]][temp]=flower_dict[x[-1]][y+"sum"]/flower_dict[x[-1]][y+"count"]
            except:
                flower_dict[x[-1]].update({temp:flower_dict[x[-1]][y+"sum"]/flower_dict[x[-1]][y+"count"]})
    return flower_dict

def flowervar(ls, flower_dict, H):
    tempsum={}
    z=-1
    for y in H:
        z+=1
        for x in ls:
            temp=y+"var"
            try:
                tempsum[x[-1]]+=(float(x[z])-flower_dict[x[-1]][y+"mean"])**2
            except:
                tempsum[x[-1]]=0
            else:
                try:
                    flower_dict[x[-1]][temp]=tempsum[x[-1]]/flower_dict[x[-1]][y+"count"]
                except:
                    flower_dict[x[-1]].update({temp:tempsum[x[-1]]/flower_dict[x[-1]][y+"count"]})
    return flower_dict

def flowerstd(ls, flower_dict, H):
    z=-1
    for y in H:
        z+=1
        for x in ls:
            temp=y+"std"
            try:
                flower_dict[x[-1]][temp]=math.sqrt(flower_dict[x[-1]][y+"var"])
            except:
                flower_dict[x[-1]].update({temp:math.sqrt(flower_dict[x[-1]][y+"var"])})
    return flower_dict

filename="iris.txt"
fulllist=[]
flowers={}
fin=open(filename,'r')
Headers = fin.readline().split(",")
Headers=Headers[0:4]
for f in fin.readlines()[0:]:
    temp=f.split(",")
    temp[-1]=temp[-1].replace("Iris-","").strip("\n")
    temp[-1]=temp[-1][0].upper()+temp[-1][1:]
    fulllist.append(temp)
print(fulllist)
flowers = flowercount(fulllist, flowers, Headers)
flowers = flowersums(fulllist,flowers, Headers)
flowers = flowermins(fulllist,flowers, Headers)
flowers = flowermax(fulllist,flowers, Headers)
flowers = flowermean(fulllist,flowers, Headers)
flowers = flowervar(fulllist,flowers, Headers)
flowers = flowerstd(fulllist,flowers, Headers)
print(flowers)
```

```python
# I imported the bible then read in the words into 2 dictionaries, 1 for word frequency and 1 for word length
# I used these to pull max and mins and other statistical values.

import statistics as st
import numpy as np
filename="bible.txt"
wordcounts={}
wordlengths={}
ls = []
fin=open(filename,'r')
for f in fin.readlines():
    ls.append(f.split())
for x in ls:
    for i in x:
        tempi=""
        i=i.lower().strip()
        for j in i:
            if j.isalpha():
                tempi += j
        i=tempi
        try:
            wordcounts[i] += 1
            wordlengths[i] = len(i)
        except:
            wordcounts[i] = 1
loop = 0
for y in reversed(sorted(wordcounts, key=wordcounts.get)):
    try:
        if loop>=10:
            raise
    except:
        break
    else:
        word = y+":"
        print(word.ljust(10),wordcounts[y])
        loop+=1
        
loop = 0
for y in reversed(sorted(wordlengths, key=wordlengths.get)):
    try:
        if loop>=10:
            raise
    except:
        break
    else:
        word = y+":"
        print(word.ljust(10),wordlengths[y])
        loop+=1
        
templen=0
templist =[]
for y in reversed(sorted(wordlengths, key=wordlengths.get)):
    templen += wordlengths[y]
    templist.append(wordlengths[y])
print("The mean word length is:", templen/len(wordlengths))
print("The Standard Deviation is: ", np.std(templist, axis=0))
```
